```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


#read csv file
data_tst = pd.read_csv('test.csv')
data_trn = pd.read_csv('train.csv')

data_test = data_tst.dropna(axis=0)
data_train = data_trn.dropna(axis=0)

#summary statistics data
data_test.describe()
data_test.columns

data_train.describe()
data_train.columns
```

```python
y_test = data_test.y
X_test = data_test.x
```

```python
y_train = data_train.y
X_train = data_train.x
```

```python
model = LinearRegression()

# Reshape X to be a 2D array
X = X_test.values.reshape(-1, 1)


# Fit the model
model.fit(X, y_test)

print("Making predictions of:")
print(X_test.head())

# Predict the data
print("Predictions are:")
y_pred = model.predict(X)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error

#evaluate the model (in-sample)
mean_absolute_error(y_test, y_pred)
```

```python
#validate the model by splitting the data

from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y_test, random_state=0)

#define the model
model = LinearRegression()

#fit the model
model.fit(X_train, y_train)

#predict the data
val_predictions = model.predict(X_val)

print(mean_absolute_error(y_val, val_predictions))

```
